```python
import requests
from IPython.display import display, Image


def show(comics_num):
    response = requests.get(f"https://xkcd.com/{comics_num}/info.0.json")
    comics = response.json()
    print(f"Title: {comics.get('title')}")
    publication_day = comics.get('day')
    if int(publication_day) < 10:
        publication_day = '0' + publication_day
    publication_month = comics.get('month')
    if int(publication_month) < 10:
        publication_month = '0' + publication_month
    print(f"Date of publication: {publication_day}.{publication_month}.{comics.get('year')}")
    display(Image(comics.get('img')))
    print(f"Transcript:\n{comics.get('transcript')}")

show(149)
print('----------------------------------------------------------------------------------------')
show(1415)
print('----------------------------------------------------------------------------------------')
show(1244)
print('----------------------------------------------------------------------------------------')
show(1365)
```
